## Single Agent Systems & Agent Pipelines

### Objective

The aim of this assignment is to build a simple single-agent AI pipeline that can understand the user's request and route it to the correct tool. The agent performs basic mathematical calculations, extracts keywords from text, and handles unsupported queries using conditional routing and structured JSON responses.

## Step 1: Import Required Libraries

In this step, the required Python libraries are imported. The `json` library is used to display outputs in a structured JSON format, while `re` is available for handling text processing if required.


```python
import re
import json
```

## Step 2: Calculator Tool

The calculator tool evaluates mathematical expressions entered by the user. It performs basic arithmetic operations and returns the computed result. If the expression is invalid, it raises an exception which will later be handled by the agent.

```python
def calculator(expression):
    """
    Evaluates a mathematical expression.
    """

    try:
        result = eval(expression)
        return result

    except Exception:
        raise ValueError("Invalid mathematical expression.")
```

## Step 3: Testing the Calculator Tool

Before integrating the calculator into the agent, it is tested with a few sample mathematical expressions to verify that it is working correctly.

```python
print(calculator("10+5"))
print(calculator("(20*5)/2"))
```

## Step 4: Keyword Extraction Tool

This tool extracts meaningful keywords from a sentence by removing common stop words such as "the", "is", and "and". It helps identify the important words present in the given text.


```python
def extract_keywords(text):

    stop_words = {
        "the","is","a","an","of","to","and",
        "in","on","for","with","this",
        "that","it","are"
    }

    words = text.lower().split()

    keywords = []

    for word in words:

        word = word.strip(".,!?")

        if word not in stop_words:

            keywords.append(word)

    return keywords
```

## Step 5: Testing the Keyword Extraction Tool

The keyword extraction function is tested with a sample sentence to ensure that it correctly removes stop words and returns only the important keywords.

```python
sentence = "Artificial Intelligence is changing the world."

print(extract_keywords(sentence))
```

## Step 6: Building the Agent

The agent acts as the decision-making component of the system. It analyzes the user's query and decides which tool should be used.

- If the query contains **"calculate"**, it is routed to the Calculator Tool.
- If the query contains **"keywords"**, it is routed to the Keyword Extraction Tool.
- If neither condition is met, the agent returns a general response.
- Any unexpected errors are handled using exception handling.

```python
def agent(query):

    query = query.strip()

    try:

        lower_query = query.lower()

        # Route to Calculator
        if "calculate" in lower_query:

            expression = lower_query.replace("calculate", "").strip()

            answer = calculator(expression)

            return {
                "type": "calculation",
                "result": answer
            }

        # Route to Keyword Extractor
        elif "keywords" in lower_query:

            text = lower_query.replace("keywords", "").strip()

            answer = extract_keywords(text)

            return {
                "type": "keywords",
                "result": answer
            }

        # General Query
        else:

            return {
                "type": "general",
                "result": "Sorry, I can only perform calculations and keyword extraction."
            }

    except Exception as e:

        return {
            "type": "error",
            "result": str(e)
        }
```

## Step 7: Testing the Agent

Different test queries are executed to verify that the routing logic works correctly. The outputs are displayed in JSON format so that the response type and result are easy to understand.

```python
print(json.dumps(agent("calculate 10+20"), indent=4))

print(json.dumps(agent("keywords Artificial Intelligence is changing healthcare"), indent=4))

print(json.dumps(agent("Hello"), indent=4))

print(json.dumps(agent("calculate 10/0"), indent=4))
```

## Step 8: Validation Test Cases

Multiple test cases are executed to check different scenarios, including valid calculations, keyword extraction, unsupported queries, and invalid mathematical expressions. This helps verify that the agent behaves correctly under different inputs.

```python
test_queries = [

    "calculate 100+25",

    "calculate (50*2)/5",

    "keywords Machine Learning is transforming education",

    "Hello",

    "calculate 5//"

]

for query in test_queries:

    print("Query:", query)

    print(json.dumps(agent(query), indent=4))

    print("-"*40)
```

## Step 9: Interactive User Input

Instead of testing one query at a time, the agent is placed inside a loop so that users can enter multiple queries continuously. The program runs until the user types **exit**.

```python
while True:

    user_query = input("Enter your query (type 'exit' to stop): ")

    if user_query.lower() == "exit":
        print("Agent Closed.")
        break

    response = agent(user_query)

    print(json.dumps(response, indent=4))
```

## Conclusion

- The agent successfully identifies the type of query using simple conditional routing.
- Mathematical expressions are processed by the calculator tool.
- Text-based queries requesting keywords are handled by the keyword extraction tool.
- Unsupported inputs receive a general response instead of causing the program to fail.
- Error handling ensures that invalid mathematical expressions do not crash the program.
- The outputs are displayed in JSON format, making them easy to read and consistent.
